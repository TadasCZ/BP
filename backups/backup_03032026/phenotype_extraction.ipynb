```python
import pandas as pd
import numpy as np
import os
```

```python
phenotypes_xlsx = "/storage/plzen4-ntis/home/tadsova/rice_data/3kRG_PhenotypeData_v20170411.xlsx"
ids_txt = "/storage/plzen4-ntis/home/tadsova/rice_data/3K_list_sra_ids.txt"
```

```python
def export_phenotype_txt(txt_path, xlsx_path, out_dir):
    df_ids = pd.read_csv(txt_path, sep="\t")
    df_ids["key"] = df_ids["Genetic_Stock_varname"].str.replace("_", " ").str.strip().str.upper()

    df_xlsx = pd.read_excel(xlsx_path, sheet_name="Phenotype Data")
    df_xlsx["key"] = df_xlsx["NAME"].str.replace("_", " ").str.strip().str.upper()

    non_pheno_cols = ["NAME", "key", "Seqno", "STOCK_ID", "GS_ACCNO", "Source_Accno", "CROPYEAR"]
    pheno_columns = []
    for col in df_xlsx.columns:
        if col not in non_pheno_cols:
            pheno_columns.append(col)

    df_all = df_xlsx.merge(
        df_ids[["key", "Genetic_Stock_varname", "3K_DNA_IRIS_UNIQUE_ID"]],
        on="key",
        how="right"
    )
    
    df_out = df_all[["3K_DNA_IRIS_UNIQUE_ID", "Genetic_Stock_varname"] + pheno_columns]

    out_path = os.path.join(out_dir, "all_phenotypes.txt")
    df_out.to_csv(out_path, sep="\t", index=False)

    print("Vytvořen soubor:", out_path)
```

```python
def phenotype_counter(file_path):
    df_all = pd.read_csv(file_path, sep="\t")
    df_pheno = df_all.iloc[:, 2:]
    phenotype_list = df_pheno.columns.tolist()
    phenotype_counts = df_pheno.count()
    sorted_counts = phenotype_counts.sort_values(ascending=False)
    
    return sorted_counts.to_dict(), phenotype_list
```

```python
export_phenotype_txt(ids_txt, phenotypes_xlsx, "phenotypes")
```

```python
counts, phenotype_list = phenotype_counter("phenotypes/all_phenotypes.txt")
if counts:
    for pheno, count in counts.items():
        print(f"{pheno}: {count} hodnot")
else:
    print("error")
```
